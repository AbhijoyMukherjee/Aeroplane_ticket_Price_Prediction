# Encoding Routes:

```python
df['Stops'] = df['Route'].str.split('?')
```

```python
length = np.zeros(13353)
for i in range(0,13353,1):
    length[i] = len(df['Stops'][i])
print(length.max())
print(length.min())
```

```python
df['First_stop']=np.zeros(13353)
df['2nd_stop']=np.zeros(13353)
for i in range(0,13353,1):
    if(length[i]==2.0):
        df['First_stop'][i] = 'None'
        df['2nd_stop'][i] = 'None'
    if(length[i]==3.0):
        df['First_stop'][i] = df['Stops'][i][1]
        df['2nd_stop'][i] = 'None'
    if(length[i]==4.0 or length[i]==5.0 or length[i]==6.0):
        df['First_stop'][i] = df['Stops'][i][1]
        df['2nd_stop'][i] = df['Stops'][i][2]
```
